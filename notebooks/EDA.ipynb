# Exploratory Data Analysis (EDA)

This notebook contains exploratory data analysis for the multi-modal emotion recognition project. We will analyze the audio, video, and text datasets to understand their characteristics and prepare for further processing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')

# Load datasets
audio_data = pd.read_csv('../data/processed/audio/audio_data.csv')
video_data = pd.read_csv('../data/processed/video/video_data.csv')
text_data = pd.read_csv('../data/processed/text/text_data.csv')

# Display the first few rows of each dataset
print('Audio Data:')
display(audio_data.head())

print('Video Data:')
display(video_data.head())

print('Text Data:')
display(text_data.head())

# Summary statistics
print('Audio Data Summary:')
display(audio_data.describe())

print('Video Data Summary:')
display(video_data.describe())

print('Text Data Summary:')
display(text_data.describe())

# Visualizations
# Example: Distribution of audio features
plt.figure(figsize=(10, 6))
sns.histplot(audio_data['feature_name'], bins=30, kde=True)
plt.title('Distribution of Audio Feature')
plt.xlabel('Feature Name')
plt.ylabel('Frequency')
plt.show()

# Further analysis can be added here...
```
